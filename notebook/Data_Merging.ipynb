## IMPORT NECESSARY LIBRARIES

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## READ ALL TABLES

```python
df_cust_fb = pd.read_csv(r'C:\Users\Administrator\OneDrive\Desktop\Case_Study\datasets\Customer_Feedback.csv')
df_MLG = pd.read_csv(r'C:\Users\Administrator\OneDrive\Desktop\Case_Study\datasets\Maintenance_Log.csv')
df_pay = pd.read_csv(r'C:\Users\Administrator\OneDrive\Desktop\Case_Study\datasets\Payments.csv')
df_resv = pd.read_csv(r'C:\Users\Administrator\OneDrive\Desktop\Case_Study\datasets\Reservations.csv')
df_tele = pd.read_csv(r'C:\Users\Administrator\OneDrive\Desktop\Case_Study\datasets\Telematics.csv')
df_fb=pd.read_csv(r'C:\Users\Administrator\OneDrive\Desktop\Case_Study\datasets\feedback.csv')
df_fb.info()
```

## REMOVE ALL DUPLICATES BEFORE MERGING

```python
# Remove Duplicate Keys 
payments = df_pay.drop_duplicates(subset="Reservation_ID")

custo_fb = df_cust_fb.drop_duplicates(subset="Reservation_ID")

maintenance = df_MLG.drop_duplicates(subset="Vehicle_ID")

telematics = df_tele.drop_duplicates(subset="Vehicle_ID")
```

## SAVE THE COPY OF MAIN TABLE "RESERVATIONS"

```python
df=df_resv.copy()
```

## MERGE ALL THE TABLES

```python
# Merge Maintenance
df = df.merge(maintenance, on="Vehicle_ID", how="left")

# Merge Telematics
df = df.merge(telematics, on="Vehicle_ID", how="left")

# Merge Payments
df = df.merge(payments, on="Reservation_ID", how="left")

# Merge Feedback
df = df.merge(custo_fb, on="Reservation_ID", how="left")
```
